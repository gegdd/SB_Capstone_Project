```python
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        d = pickle.load(fo, encoding='bytes')
        # decode utf8
        d_decoded = {}
        for k, v in d.items():
            d_decoded[k.decode('utf8')] = v
        d = d_decoded
    return d

def load_batch(file):
    d = unpickle(file)
    data = d['data']
    labels = d['labels']
    #data = data.reshape(data.shape[0], 32, 32, 3)
    return data, labels

def stack_batches():
    
    y_train = []
    
    for i in range(1,6):
        x_train_temp, y_train_temp = load_batch('data_batch_' + str(i))
        y_train += y_train_temp
        
        if i==1:
            x_train = x_train_temp
        else:
            x_train = np.row_stack([x_train, x_train_temp])
            
    x_test, y_test = load_batch('test_batch')
    
    y_train = np.array(y_train)
    y_test = np.array(y_test)
    
    return x_train, y_train, x_test, y_test
```

```python
x_train,y_train,x_test,y_test = stack_batches()
```
